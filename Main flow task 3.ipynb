```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:\\Users\\Laptop\\Downloads\\Telegram Desktop\\train (3).csv")
```

```python
data.head(5)
```

```python
#scatter plot
plt.scatter(data['Survived'],data['Age'])
            
#adding title to the plot
plt.title('Scatter plot')
            
#labels
plt.xlabel('Survived')
plt.ylabel('Age')

#show result
plt.show()
```

```python
#line chart
plt.plot(data['Survived'],data['Age'])
            
#adding title to the plot
plt.title('line chart')
            
#labels
plt.xlabel('Survived')
plt.ylabel('Age')

#show result
plt.show()
```

```python
#Bar chart
plt.bar(data['Survived'],data['Age'])
            
#adding title to the plot
plt.title('Bar plot')
            
#labels
plt.xlabel('Survived')
plt.ylabel('Age')

#show result
plt.show()
```

```python
#histogram
#Histogram
plt.hist(data['Survived'])

#adding title to the plot
plt.title("Histogram")

#Showing result
plt.show()
```
